Measuring Limit of Detection
LOD = 3.3 x SD of the blank / Calibration Curve

```python
#DataFrames
rows_to_exclude = [0, 1, 2, 3, 4]
Correl_0 = pd.read_csv("Day 0 Correl.csv")
Correl_1 = pd.read_csv("Day 1 Correl.csv")
Correl_2 = pd.read_csv("Day 2 Correl.csv")
Correl_3 = pd.read_csv("Day 3 Correl.csv")
Correl_5 = pd.read_csv("Day 5 Correl.csv")
Correl_6 = pd.read_csv("Day 6 Correl.csv")
SerCorrel_0 = Correl_0.drop(rows_to_exclude, axis = 0)
SerCorrel_1 = Correl_1.drop(rows_to_exclude, axis = 0)
SerCorrel_2 = Correl_2.drop(rows_to_exclude, axis = 0)
SerCorrel_3 = Correl_3.drop(rows_to_exclude, axis = 0)
SerCorrel_5 = Correl_5.drop(rows_to_exclude, axis = 0)
SerCorrel_6 = Correl_6.drop(rows_to_exclude, axis = 0)
```

```python
#Measuring LOD for Day 0
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

#find the standard deviation of the blank
Blank0 = np.array([0.047, 0.05, 0.056, 0.053, 0.052])
Blank0SD = np.std(Blank1)
print(Blank1SD)

#Find the linear regression line
plt.plot(SerCorrel_0 ["Dilution"], SerCorrel_0 ["OD600"], "o")
plt.xlabel("Dilutions")
plt.ylabel("OD600")
plt.title("Calibration Curves of OD600 of Day 0")
reg = stats.linregress(SerCorrel_0 ["Dilution"], SerCorrel_0 ["OD600"])
plt.plot(SerCorrel_0 ["Dilution"], reg.intercept + reg.slope*SerCorrel_0 ["Dilution"], color = "red")
Day0slope = reg.slope
Day0intercept = reg.intercept
print("slope:", Day0slope)
print("intercept:", Day0intercept)

def LOD(regressionline, standard_dev):
    detection = (3.3*(standard_dev))/(regressionline)
    return detection

Day0LOD = LOD(Day0slope, Blank0SD)
print(Day0LOD)

predict_value = Day0slope * SerCorrel_0["Dilution"] + Day0intercept
SS_res = np.sum(((SerCorrel_0["OD600"] - predict_value)**2))

Weightmean = np.mean(SerCorrel_0["OD600"])
SS_tot = np.sum((SerCorrel_0["OD600"] - Weightmean)**2)

R_sq = 1 - (SS_res/SS_tot)
print("R^2:", R_sq)
```
